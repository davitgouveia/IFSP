# Exercício 1.a

### Importando biblioteca e criando a função

```python
import numpy as np

# Definindo a função
def calculate_entropy(imagem):
    n = 10  # Número de símbolos utilizados

    # bincount calcula o número de ocorrencias de um valor na array
    histogram = np.bincount(imagem.flatten())

    # Calcula as probabilidades dos símbolos
    probabilities = histogram / float(imagem.size) # tamanho da array

    entropy = 0.0
    # percorre as probabilidades e realizar calcula a entropia
    for p in probabilities:
        entropy -= p * np.log2(p)  # Calcula a entropia usando a fórmula

    return entropy
```
