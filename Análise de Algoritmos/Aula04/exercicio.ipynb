# Exercício Aula 4

Algoritmo A: T(n) = n(n+5)

Algoritmo B: T(n) = 10n+10

## Comparação gráfica

```python
import matplotlib.pyplot as plt
```

```python
def countInstructionsA(n):
    instructionsTotal = n*(n + 5)
    return instructionsTotal

def countInstructionsB(n):
    instructionsTotal = 10*n + 10
    return instructionsTotal
```

```python
instructions_countA = [countInstructionsA(n) for n in range(8)]
instructions_countB = [countInstructionsB(n) for n in range(8)]

plt.plot(instructions_countA , 'r')
plt.plot(instructions_countB , 'g')
plt.xlabel('Valor de N')
plt.ylabel('Contagem de instruções')
plt.grid()
plt.show()
```

### Algorítmo A = Vermelho
### Algorítmo B = Verde

Com a análise, é possível perceber que o Algorítmo B é o mais eficiente até o valor de instruções for de aproximadamente 6,5. Após esse valor, o Algorítmo A passa a ser o mais eficiente.
